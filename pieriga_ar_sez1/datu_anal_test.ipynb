```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('2024_gg.csv', sep=';')
```

```python
# rindu un kolonu skaits
data.shape
```

```python
# Datu tipi
data.dtypes
```

```python
# Vai ir kādas tukšas vērtības
data.isnull().sum()
```

```python
data.describe()
```

```python
corr = data.corr()
corr.shape
```

```python
plt.figure(figsize=(20, 20))
sns.heatmap(corr, annot=True, fmt='.2f', annot_kws={'size':16}, cmap='coolwarm')
```

```python
plt.figure(figsize=(7, 5))
sns.histplot(data=data, x='Darījuma summa, EUR', kde=True)
plt.ylabel('Skaits')
plt.show()
```

```python
# Pielieoto logaritmisko tranformaciju
data['Darījuma summa, EUR'] = np.log1p(data['Darījuma summa, EUR'])
```

```python
plt.figure(figsize=(7, 4))
sns.histplot(data=data, x='Darījuma summa, EUR', kde=True)
plt.ylabel('Skaits')
plt.show()
```
